```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

```

```python
categorical_features = pd.read_csv('categorical.csv')
display(categorical_features)

numerical_features = pd.read_csv('numerical.csv')
print(numerical_features.columns.tolist())

X = pd.concat([categorical_features, numerical_features], axis = 1)


targets = pd.read_csv('target.csv')
y = targets['TARGET_B']

data = pd.concat([X, y], axis = 1)
print(data.shape)
data_no_nan = data.dropna()
print(data_no_nan.shape)


X = data_no_nan.drop('TARGET_B', axis = 1)
y = data_no_nan['TARGET_B']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1337)
```

```python
X_train_numerical = X_train.select_dtypes(np.number)
X_test_numerical = X_test.select_dtypes(np.number)

X_train_categorical = X_train.select_dtypes(object)
X_test_categorical = X_test.select_dtypes(object)

display(X_test_numerical)

```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler().fit(X_train_numerical)

X_train_numerical_scaled = scaler.transform(X_train_numerical)
X_train_numerical_scaled_df = pd.DataFrame(X_train_numerical_scaled, columns=scaler.get_feature_names_out(X_train_numerical.columns))

X_test_numerical_scaled = scaler.transform(X_test_numerical)
X_test_numerical_scaled_df = pd.DataFrame(X_test_numerical_scaled, columns=scaler.get_feature_names_out(X_test_numerical.columns))
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first')
encoder.fit(X_train_categorical)

X_train_categorical_encoded = encoder.transform(X_train_categorical)
X_train_categorical_encoded_df = pd.DataFrame(X_train_categorical_encoded.toarray(), columns=encoder.get_feature_names_out(X_train_categorical.columns))

X_test_categorical_encoded = encoder.transform(X_test_categorical)
X_test_categorical_encoded_df = pd.DataFrame(X_test_categorical_encoded.toarray(), columns=encoder.get_feature_names_out(X_test_categorical.columns))
```

```python
X_train_categorical_encoded
```

```python
X_train = pd.concat([X_train_categorical_encoded_df, X_train_numerical_scaled_df], axis = 1)
X_train

X_test = pd.concat([X_test_categorical_encoded_df, X_test_numerical_scaled_df], axis = 1)
X_test
```

```python

y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

models = ['LogisticRegression', 'KNeighborsClassifier', 'DecisionTreeClassifier']
```

```python
def train_test_all_models(X, X_test, y, y_test, models):
    counter = 0
    for model in models: 
        
        
        if model == 'LogisticRegression':
            current_model = LogisticRegression(max_iter = 5000) # create model 
            current_model.fit(X, y) # fit model
        else:
            current_model = eval(model + "()") # create model 
            current_model.fit(X, y) # fit model
        
        y_predicted = current_model.predict(X)
        y_test_predicted = current_model.predict(X_test)
        
        
        from sklearn.metrics import mean_squared_error as mse
        from sklearn.metrics import mean_absolute_error as mae
        
        r2 = round(current_model.score(X, y), 3)
        
        print (f'Model {models[counter]}:')
        print ('The train r2 score is: {}.'.format(r2))

        r2 = round(current_model.score(X_test, y_test), 3)

        print ('The test r2 score is: {}.'.format(r2))
        print ('\n')
        counter += 1 
        
```

```python
train_test_all_models(X_train, X_test, y_train, y_test, models)
```

```python
# Model LogisticRegression:
# The train r2 score is: 0.949.
# The test r2 score is: 0.948.


# Model KNeighborsClassifier:
# The train r2 score is: 0.95.
# The test r2 score is: 0.947.


# Model DecisionTreeClassifier:
# The train r2 score is: 1.0.
# The test r2 score is: 0.889.
```

```python
imbalance_data = pd.concat([X_train, y_train], axis = 1)

imbalance_data['TARGET_B'].value_counts()

```

```python
from sklearn.utils import resample

category_0 = imbalance_data[imbalance_data['TARGET_B'] == 0]
category_1 = imbalance_data[imbalance_data['TARGET_B'] == 1]

category_0_undersampled = resample(category_0,
                                   replace=False,
                                   n_samples = len(category_1))

data_downsampled = pd.concat([category_0_undersampled, category_1], axis=0)

data_downsampled['TARGET_B'].value_counts()
```

```python
X_downsampled = data_downsampled.drop('TARGET_B', axis = 1)
y_downsampled = data_downsampled['TARGET_B']
```

```python
train_test_all_models(X_downsampled, X_test, y_downsampled, y_test, models)
```

```python
# Model LogisticRegression:
# The train r2 score is: 0.62.
# The test r2 score is: 0.592.


# Model KNeighborsClassifier:
# The train r2 score is: 0.706.
# The test r2 score is: 0.532.


# Model DecisionTreeClassifier:
# The train r2 score is: 1.0.
# The test r2 score is: 0.514.
```

```python
category_1_oversampled = resample(category_1,
                                  replace=True,
                                  n_samples = len(category_0))

data_upsampled = pd.concat([category_0, category_1_oversampled], axis=0)

data_upsampled['TARGET_B'].value_counts()
```

```python
X_upsampled= data_upsampled.drop('TARGET_B', axis = 1)
y_upsampled = data_upsampled['TARGET_B']
```
